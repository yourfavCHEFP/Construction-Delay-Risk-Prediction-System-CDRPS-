```python
import sys
from pathlib import Path

import pandas as pd

# Ensure project root is importable when running from notebooks/
sys.path.append(str(Path.cwd().resolve().parent))

from src.ingestion.file_loader import load_file, preview_data
from src.validation.schema_validator import run_validation
```

```python
# Centralized notebook configuration
DATASET_PATH = "../../archive(1)/Road Constuction Delay Survey.csv"
PREVIEW_ROWS = 5

# Expanded required schema based on dataset header
REQUIRED_COLUMNS = [
    "Respondent Number ",
    "Category 1: Materials",
    "Category 1: Materials.1",
    "Category 1: Materials.2",
    "Category 1: Materials.3",
    "Category 1: Materials.4",
    "Category 2: Labor and Equipment",
    "Category 2: Labor and Equipment.1",
    "Category 2: Labor and Equipment.2",
    "Category 2: Labor and Equipment.3",
    "Category 2: Labor and Equipment.4",
    "Category 2: Labor and Equipment.5",
    "Category 2: Labor and Equipment.6",
    "Category 2: Labor and Equipment.7",
    "Category 3: Financing",
    "Category 3: Financing.1",
    "Category 3: Financing.2",
    "Category 3: Financing.3",
    "Category 3: Financing.4",
    "Category 4: Design and Documentation",
    "Category 4: Design and Documentation.1",
    "Category 4: Design and Documentation.2",
    "Category 4: Design and Documentation.3",
    "Category 4: Design and Documentation.4",
    "Category 4: Design and Documentation.5",
    "Category 5: Management and Organization ",
    "Category 5: Management and Organization .1",
    "Category 5: Management and Organization .2",
    "Category 5: Management and Organization .3",
    "Category 5: Management and Organization .4",
    "Category 5: Management and Organization .5",
    "Category 6: Schedule ",
    "Category 6: Schedule .1",
    "Category 7: Contractual issues ",
    "Category 7: Contractual issues .1",
    "Category 7: Contractual issues .2",
    "Category 7: Contractual issues .3",
    "Category 7: Contractual issues .4",
    "Category 8: Scope of work",
    "Category 8: Scope of work.1",
    "Category 8: Scope of work.2",
    "Category 8: Scope of work.3",
    "Category 9: External issues ",
    "Category 9: External issues .1",
    "Category 9: External issues .2",
    "Respondant Information",
    "Respondant Information.1",
    "Respondant Information.2",
    "Respondant Information.3",
    "Respondant Information.4",
    "Respondant Information.5",
]
```

```python
df_raw = load_file(DATASET_PATH)

print(f"Dataset loaded from: {DATASET_PATH}")
print(f"Shape: {df_raw.shape}")
print("Columns:")
print(df_raw.columns.tolist())

print("\nPreview:")
preview_data(df_raw, PREVIEW_ROWS)
```

```python
validation_result = run_validation(df_raw, REQUIRED_COLUMNS)

print("missing_columns:")
print(validation_result["missing_columns"])

print("\nmissing_values_summary:")
print(validation_result["missing_values_summary"])

print("\ninconsistencies:")
print(validation_result["inconsistencies"])

print("\nhas_critical_issues:")
print(validation_result["has_critical_issues"])
```

```python
print("CHECK THIS OUT: Validation Report")
print("=" * 50)

if validation_result["has_critical_issues"]:
    print("Critical validation failure - pipeline stopped.")
    print("Missing required columns:")
    print(validation_result["missing_columns"])
else:
    has_warnings = bool(validation_result["missing_values_summary"]) or bool(validation_result["inconsistencies"])

    if has_warnings:
        print("Validation passed with warnings - you may proceed to preparation.")
        print("Missing values summary:")
        print(validation_result["missing_values_summary"])
        print("Inconsistencies:")
        print(validation_result["inconsistencies"])
    else:
        print("Validation successful - dataset is ready for preparation.")
```

```python
if validation_result["has_critical_issues"]:
    print("STOP: Critical issues detected. Preparation stage is blocked.")
else:
    print("OK: Validation gate passed. You can proceed to preparation in Notebook 02.")
```
