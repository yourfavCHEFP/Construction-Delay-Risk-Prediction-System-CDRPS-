# Pipeline Overview

This notebook implements the full data preparation pipeline for the Construction Delay Risk Prediction System. The goal of this stage is to transform the raw dataset into a clean, structured, and machine‑learning‑ready format. The pipeline follows a modular architecture, where each step is handled by a dedicated component within the `src` directory. This ensures clarity, maintainability, and reproducibility across the entire workflow.

The preparation pipeline consists of the following stages:

# 1. Data Cleaning
The dataset is first passed through the cleaning module, which:
- Fixes incorrect or inconsistent data types  
- Handles missing values using transparent rules  
- Removes or flags invalid rows (e.g., negative durations or impossible values)  
- Ensures the dataset is structurally sound before further processing  

This step guarantees that the data entering the model is reliable and free from basic quality issues.

# 2. Categorical Encoding
Categorical features are transformed into numerical representations using Label Encoding.  
This step:
- Converts string‑based categories into integer labels  
- Stores the fitted encoders for later use during inference  
- Ensures consistency between training and deployment environments  

# 3. Numerical Scaling
Numerical features are standardized using a scaling technique to ensure:
- All numeric columns share a similar range  
- The model is not biased toward features with larger magnitudes  
- Training becomes more stable and efficient  

The fitted scaler is also preserved for use during prediction.

# 4. Dataset Splitting
The cleaned and transformed dataset is divided into:
- **Training set
- **Validation set
- **Test set

This split supports:
- Model training  
- Hyperparameter tuning  
- Final unbiased evaluation  

Optional stratification ensures that the target variable maintains its distribution across all splits.
