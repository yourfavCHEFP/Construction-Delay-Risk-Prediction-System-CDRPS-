The goal is to ensure that the dataset meets all structural and type requirements before it enters the preparation pipeline.

Running validation at this stage ensures:

- All required columns exist  
- Column data types match expectations  
- Missing values are detected early

If validation fails, the errors displayed here must be corrected before continuing.

```python
if not validation_results["has_critical_issues"]:
    print("Validation passed. Dataset is structurally sound.")
else:
    print("Validation failed. Issues detected:")

    if validation_results["missing_columns"]:
        print("\nMissing Columns:")
        for col in validation_results["missing_columns"]:
            print("-", col)

    if validation_results["invalid_types"]:
        print("\nInvalid Data Types:")
        for issue in validation_results["invalid_types"]:
            print("-", issue)

    if validation_results["inconsistencies"]:
        print("\nInconsistencies Found:")
        for issue in validation_results["inconsistencies"]:
            print("-", issue)

    if validation_results["missing_values_summary"]:
        print("\nMissing Values Summary:")
        for col, count in validation_results["missing_values_summary"].items():
            print(f"- {col}: {count} missing")
```
